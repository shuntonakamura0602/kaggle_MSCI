# merge csv

```python
CITE_PATH = f"/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/cite_sub/{CITE_NAME}.csv"
MULTI_PATH = f"/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/multi_sub/{MULTI_NAME}.csv"
SUB_NAME = f"{CITE_NAME}+{MULTI_NAME}.csv"
```

```python
cite_submission = pd.read_csv(CITE_PATH)
cite_submission = cite_submission.set_index("row_id")
cite_submission = cite_submission["target"]
```

```python
multi_submission = pd.read_csv(MULTI_PATH)
multi_submission = multi_submission.set_index("row_id")
```

```python
multi_submission.target[multi_submission.target.isnull()] = cite_submission[multi_submission.target.isnull()]
```

```python
multi_submission.isnull().any()
```

# to_csv

```python
multi_submission.to_csv(SUB_NAME)
```
