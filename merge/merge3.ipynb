# merge csv

```python
CITE_PATH = "/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/cite_sub/keras1.csv"
MULTI_PATH = "/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/multi_sub/ridge2.csv"
SUB_PATH = "/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/submission/keras1+ridge2.csv"
```

```python
cite_submission = pd.read_csv(CITE_PATH)
cite_submission = cite_submission.set_index("row_id")
cite_submission = cite_submission["target"]
```

```python
multi_submission = pd.read_csv(MULTI_PATH)
multi_submission = multi_submission.set_index("row_id")
```

```python
multi_submission.target[multi_submission.target.isnull()] = cite_submission[multi_submission.target.isnull()]
```

# to_csv

```python
multi_submission.to_csv(SUB_PATH)
```
