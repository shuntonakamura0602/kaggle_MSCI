```python
import os
DATA_DIR = "/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset"
FP_MULTIOME_TRAIN_TARGETS = os.path.join(DATA_DIR,"train_multi_targets_values.sparse.npz")
FP_MULTIOME_TRAIN_INPUTS = os.path.join(DATA_DIR,"train_multi_inputs_values.sparse.npz")
FP_MULTIOME_TEST_INPUTS = os.path.join(DATA_DIR,"test_multi_inputs_values.sparse.npz")
```

```python
%%time
train_inputs = scipy.sparse.load_npz(FP_MULTIOME_TRAIN_INPUTS)
```

```python
%%time
train_targets = scipy.sparse.load_npz(FP_MULTIOME_TRAIN_TARGETS)
```

```python
%%time
pca = TruncatedSVD(n_components=N_COMPONENTS, random_state=42)
train_inputs = pca.fit_transform(train_inputs)
```

```python
%%time
pca2 = TruncatedSVD(n_components=N_COMPONENTS, random_state=42)
train_targets = pca2.fit_transform(train_targets)
```
