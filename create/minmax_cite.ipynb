```python
import pandas as pd
import numpy as np
import scipy.sparse
from sklearn.preprocessing import MinMaxScaler
```

# conversion functions

```python
original_train = pd.read_hdf("/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset/train_cite_inputs.h5")
```

```python
original_test = pd.read_hdf('/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset/test_cite_inputs.h5')
```

```python
original_train.shape,original_test.shape
```

```python
original_all = pd.concat([original_train,original_test])
original_all.shape
```

```python
original_all.head()
```

```python
scaler = MinMaxScaler()
original_all = scaler.fit_transform(original_all)
```

```python
original_train = original_all[:70988]
original_test = original_all[70988:]
```

```python
original_train.shape,original_test.shape
```

```python
original_train = pd.DataFrame(data=original_train,columns=original_all.columns)
```

```python
original_test = pd.DataFrame(data=original_test,columns=original_all.columns)
```

```python
original_train.to_hdf("/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset/minmax_train_cite_inputs.h5",'key',mode='w')
```
