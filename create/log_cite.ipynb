# conversion functions

```python
original_train = pd.read_hdf("/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset/train_cite_inputs.h5")
```

```python
original_test = pd.read_hdf('/content/drive/MyDrive/Colab Notebooks/kaggle/MSCI/dataset/test_cite_inputs.h5')
```

```python
original_train.shape,original_test.shape
```

```python
original_all = pd.concat([original_train,original_test])
original_all.shape
```

```python
original_col = original_all.columns
```

```python
#scaler = StandardScaler()
original_all = np.log1p(original_all)
```

```python
original_train = original_all[:70988]
original_test = original_all[70988:]
```

```python
original_train.shape,original_test.shape
```

```python
original_train = pd.DataFrame(data=original_train,columns=original_col)
```
